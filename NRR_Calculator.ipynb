# NRR CALCULATOR
## Here to make your work easier

```python
matches = int(input("Enter number of matches: "))
runs_scored = list(map(int,input("Enter runs scored: ").split(',')))
overs_faced = list(map(float,input("Enter overs faced: ").split(',')))
wickets_lost = list(map(int,input("Enter wickets lost: ").split(',')))
runs_conceded = list(map(int,input("Enter runs conceded: ").split(',')))
overs_bowled = list(map(float,input("Enter overs bowled: ").split(',')))
wickets_taken = list(map(int,input("Enter wickets taken: ").split(',')))

def overs_to_balls(over):
    whole = int(over)
    balls = round((over - whole) * 10)
    return whole * 6 + balls

total_runs_scored = sum(runs_scored)
total_runs_conceded = sum(runs_conceded)

total_balls_faced = 0
total_balls_bowled = 0

for i in range(matches):
    if wickets_lost[i] == 10:
        total_balls_faced += 120 
    else:
        total_balls_faced += overs_to_balls(overs_faced[i])

    if wickets_taken[i] == 10:
        total_balls_bowled += 120 
    else:
        total_balls_bowled += overs_to_balls(overs_bowled[i])

run_rate_scored = (total_runs_scored / total_balls_faced) * 6
run_rate_conceded = (total_runs_conceded / total_balls_bowled) * 6

nrr = run_rate_scored - run_rate_conceded

print(f"NRR: {round(nrr,3)}")
```
